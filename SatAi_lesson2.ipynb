
## Pandas - Usando dataframes

Os daré una breve introducción, pero recomiendo encareciadamente hacer el cursillo  [10 Minutes to pandas](http://pandas.pydata.org/pandas-docs/version/0.15/10min.html)

```python
import numpy as np #librería básica. Contiene funciones, numeros como pi, e, y muchas cosas más 
```

La librería de pandas es muy útil para archivos muy grandes, que en excel serían imposibles o muy pesados de manejar.

Puede realizar consultas y muchas operaciones sobre la tabla/base de datos. 
Sus objetos básicos son las `Series` y los `DataFrame`. 

Vamos a realizar un pequeño ejemplo. Construiremos nuestro propio DataFrame para ir entrando en temario. Definimos la lista de alumnos

###Ejemplo  alumnosCSV

```python
df2 = pd.read_csv(path) 
df2
```

Dividimos mediante el carater `@` el string para mayor claridad

```python
df2['Name'] = [alum.split('@')[0] for alum in df2.Email]
df2['Org'] = [alum.split('@')[1] for alum in df2.Email]
df2
```

```python
nPerG = 6 # numero de personas por grupo 
nGroups = int(np.ceil(df2.shape[0]/nPerG)) # numero de grupos
nGroups
```

Ya que sabemos que son 5 grupos, vamos a darle un nombre

```python
lstTeamsNames= ['Agua', 'Fuego', 'Hielo', 'Tierra', 'Rayo']
```

Multiplicamos esta lista

```python
lstTeams = lstTeamsNames* nPerG
lstTeams
```

Hay que distinguir en python, los metodos que actuan sobre un objeto, y las que devuelven un objeto.  ne este caso ramdom.shuffle actua sobre el objeto argumento y no devuelve nada. 

```python
np.random.shuffle(lstTeams)
lstTeams
```

```python
df2['Team'] = lstTeams
df2
```

Podemos filtrar rápidamente así por equipo. 

```python
df2[df2.Team == 'Agua']
```

Y podemos reordenar nuestro dataframe por equipos.

```python
df2 = df2.sort_values(by = ['Team'])
df2
```

Podemos eliminar las columnas que no nos interesan

```python
df2.drop(columns=['Email', 'Org'])
```

Y podemos salvar nuestro trabajo

```python
name = 'foo_listaEquipos.csv'
folder = '/' + 'content/gdrive/My Drive' + '/'
path = folder + name 
df2.to_csv( path )
```
